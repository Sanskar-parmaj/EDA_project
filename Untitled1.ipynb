```python
def remove_outlier(col) :
  sorted(col)
  Q1,Q3=col.quantile([0.25,0.75])
  IQR=Q3-Q1
  lower_range=Q1-(1.5*IQR)
  upper_range=Q3+(1.5*IQR)
  return lower_range, upper_range
```

```python
lrtotal,urtotal=remove_outlier(df['total_goals'])

df['total_goals']=np.where(df['total_goals']>urtotal,urtotal,df['total_goals'])
df['total_goals']=np.where(df['total_goals']<lrtotal,lrtotal,df['total_goals'])

df.shape
# example
```

Replacing NULL values in Catrgorical Columns using mode

```python
mode = df['referee'].mode()[0]

df['referee'].replace(np.nan,mode,inplace=True)
```

Replacing NULL values in Catrgorical Columns using mode

```python
median1 = df['halftime_away'].median()
median2 = df['halftime_home'].median()

df['halftime_away'].replace(np.nan,median1,inplace=True)
df['halftime_home'].replace(np.nan,median2,inplace=True)
```
